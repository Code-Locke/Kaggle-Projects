## 1. Overview 
For the dataset, I used PDA to decrease the dimensionality from 16 to 2, and used the KNN classifier to classify the dataset. 

## 2. Import
We are going to use only pandas, numpy and matplotlib for this exercise and implement all algorithms from scratch.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 3. PCA
### 3.1. Loading data
We are going to load and check the dataset structure

```python
file_path = '/kaggle/input/dry-bean-dataset/Dry_Bean_Dataset/Dry_Bean_Dataset.xlsx'
sh_name = 'Dry_Beans_Dataset'
df = pd.read_excel(file_path, sheet_name=sh_name)

print("Dataset shape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nFirst few rows:")
print(df.head())

features = df.iloc[:, :16].values  
bean_type = df.iloc[:, 16].values

print(f"\nFeature matrix shape: {features.shape}")
print(f"Labels shape: {bean_type.shape}")
```

### 3.2. Standardization

Before getting into the actual PCA algorithm, since the data has different scales for different labels, we will standardize the data.

```python
n_components = 2

mean = np.mean(features, axis=0)
sig = np.std(features, axis=0)
x_centre = features-mean
x_stand = x_centre / sig

print(f"\nData standardized. Shape: {x_stand.shape}")
```

### 3.3. Covariance and Eigenvalue

We will compute the covariance matrix of the standardized data, which will describe how each feature varies with other features and since PCA needs eigendecomposition, we will compute the eigenvectors and the eigenvalues

```python

n_samples = features.shape[0]
cov_matrix = np.dot(x_stand.T, x_stand) / (n_samples - 1)

print(f"Covariance matrix shape: {cov_matrix.shape}")

e_val, e_vec = np.linalg.eig(cov_matrix)

print(f"Eigenvalues shape: {e_val.shape}")
print(f"Eigenvectors shape: {e_vec.shape}")
```

### 3.4. Eigenvalues and Eigenvector sorting
We will then sort eigenvectors by their eigenvalues in descending order and select the top components, which is in this case, 2.

After that, we project the standardized data. Also, we will show the reduced data shape and the Information Retention Ratio, and the total variance retained.

```python
idx = np.argsort(e_val)[::-1]
e_val_sorted = e_val[idx]
e_vec_sorted = e_vec[:, idx]

components = e_vec_sorted[:, :n_components]
selected_eigenvalues = e_val_sorted[:n_components]

print(f"\nSelected {n_components} principal components")
print(f"Components shape: {components.shape}")

reduced_feat = np.dot(x_stand, components)

print(f"Reduced data shape: {reduced_feat.shape}")

info_ratio = selected_eigenvalues / np.sum(e_val)
print(f"\Information Retention ratio: {info_ratio}")
print(f"Total variance retained: {np.sum(info_ratio):.2%}")
```

### 3.5. Plotting

We will plot the 2D data created after reducing the dimensionality, and show the scatter plot for bean types. (Different types shown by different colors.). Also, we will

```python
plt.figure(figsize=(12, 8))

scatter = plt.scatter(reduced_feat[:, 0], reduced_feat[:, 1], 
                     c=pd.factorize(bean_type)[0], 
                     cmap='viridis', 
                     alpha=0.6,
                     s=20)

plt.xlabel(f'PC1 ({info_ratio[0]:.2%} variance)')
plt.ylabel(f'PC2 ({info_ratio[1]:.2%} variance)')
plt.title('PCA of Bean Dataset - 2D Projection')
plt.colorbar(scatter, label='Bean Type')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


all_var_ratio = e_val_sorted / np.sum(e_val)
cum_var = np.cumsum(all_var_ratio)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))


ax1.bar(range(1, len(all_var_ratio) + 1), all_var_ratio)
ax1.set_xlabel('Principal Component')
ax1.set_ylabel('Information Retention Ratio')
ax1.set_title('Information Retention by Each Component')
ax1.grid(True, alpha=0.3)

ax2.plot(range(1, len(cum_var) + 1), cum_var, 'bo-')
ax2.axhline(y=0.95, color='r', linestyle='--', label='95% variance')
ax2.set_xlabel('Number of Components')
ax2.set_ylabel('Cumulative Variance Explained')
ax2.set_title('Cumulative Variance Explained')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nNumber of components needed for 95% variance: {np.argmax(cum_var >= 0.95) + 1}")
```

## 4. KNN Classifier

### 4.1. Data Split

As usual with other common practices, we will seed the random as 42, and split the training and testing data into 80/20.

```python
np.random.seed(42)
n_sample_total = reduced_feat.shape[0]
indices = np.random.permutation(n_sample_total)
split_idx = int(0.8 * n_sample_total)

train_indices = indices[:split_idx]
test_indices = indices[split_idx:]

x_train = reduced_feat[train_indices]
y_train = bean_type[train_indices]
x_test = reduced_feat[test_indices]
y_test = bean_type[test_indices]

print(f"\nTraining set size: {x_train.shape[0]}")
print(f"Testing set size: {x_test.shape[0]}")
```

### 4.2. KNN Implementation

We will define the euclidean distance since we will be using this equation a lot.
After that, we will iterate through the sample, compute the distance to every training sample, sort them by the distance, select the 'K Nearest Neighbors' and choose the predicted label.

```python

def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

def knn_predict(X_train, y_train, x_test, k):
    
    distances = []
    for i in range(len(X_train)):
        dist = euclidean_distance(x_test, X_train[i])
        distances.append((dist, y_train[i]))
    
    distances.sort(key=lambda x: x[0])
    k_nearest = distances[:k]
    k_nearest_labels = [label for (_, label) in k_nearest]
    
    unique_labels, counts = np.unique(k_nearest_labels, return_counts=True)
    return unique_labels[np.argmax(counts)]

def knn_predict_all(X_train, y_train, X_test, k):
    predictions = []
    for i in range(len(X_test)):
        pred = knn_predict(X_train, y_train, X_test[i], k)
        predictions.append(pred)
    return np.array(predictions)
```

### 4.3. Train & Testing KNN

We will use the previously split training set and the test set here. The training set will be used to find nearest neighbors,and the test set will be used to evaluate performance. 

We will test KNN for several values of k, since we do not know the optimal k value that will give us the most accurate data that is not overfitted.

After that, we will calculate the accuracy and record the most accurate(best) k for future use. However, since we reduced the dimensionality quite a lot, we will not get a 95% accuracy.

```python

k_values = [3, 5, 7, 9, 11, 13, 15]

print("\nTesting KNN with different k values...")
print("-" * 60)

best_k = None
best_accuracy = 0

for k in k_values:
    print(f"\nTesting with k={k}...")
    y_pred = knn_predict_all(x_train, y_train, x_test, k)
    
    accuracy = np.sum(y_pred == y_test) / len(y_test)
    
    print(f"Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

print("\n" + "="*60)
print(f"BEST RESULT: k={best_k} with accuracy={best_accuracy:.4f} ({best_accuracy*100:.2f}%)")
print("="*60)
```

### 4.4. Visualization

We will use a confusion matrix with true labels as rows, and predictions as columns. Each cell in the matrix will show the number of samples classified, thus showing us the performance on a class basis.

After that, we will also compute the class accuracy for all 7 classes, the number of wrong predictions, and some samples of wrong predictions to show if there are any, patterns in the wrong predictions.

And to visualize that further, we will use scatterplots to compare class boundaries.

```python

y_pred_final = knn_predict_all(x_train, y_train, x_test, best_k)


unique_classes = np.unique(bean_type)
n_classes = len(unique_classes)
confusion_matrix = np.zeros((n_classes, n_classes), dtype=int)

class_to_idx = {cls: idx for idx, cls in enumerate(unique_classes)}

for true_label, pred_label in zip(y_test, y_pred_final):
    true_idx = class_to_idx[true_label]
    pred_idx = class_to_idx[pred_label]
    confusion_matrix[true_idx, pred_idx] += 1


fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(confusion_matrix, cmap='Blues')

cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Number of Predictions', rotation=270, labelpad=20)

ax.set_xticks(np.arange(n_classes))
ax.set_yticks(np.arange(n_classes))
ax.set_xticklabels(unique_classes)
ax.set_yticklabels(unique_classes)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

for i in range(n_classes):
    for j in range(n_classes):
        text = ax.text(j, i, confusion_matrix[i, j],
                      ha="center", va="center", color="black" if confusion_matrix[i, j] < confusion_matrix.max()/2 else "white")

ax.set_xlabel('Predicted Label')
ax.set_ylabel('True Label')
ax.set_title(f'Confusion Matrix (KNN with k={best_k})')
plt.tight_layout()
plt.show()

print("\nPer-class accuracy:")
print("-" * 40)
for idx, cls in enumerate(unique_classes):
    class_correct = confusion_matrix[idx, idx]
    class_total = np.sum(confusion_matrix[idx, :])
    class_accuracy = class_correct / class_total if class_total > 0 else 0
    print(f"{cls:15s}: {class_accuracy:.4f} ({class_accuracy*100:.2f}%)")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

scatter1 = ax1.scatter(x_test[:, 0], x_test[:, 1], 
                       c=pd.factorize(y_test)[0], 
                       cmap='viridis', 
                       alpha=0.6, s=30)
ax1.set_xlabel('PC1')
ax1.set_ylabel('PC2')
ax1.set_title('True Labels (Test Set)')
ax1.grid(True, alpha=0.3)

scatter2 = ax2.scatter(x_test[:, 0], x_test[:, 1], 
                       c=pd.factorize(y_pred_final)[0], 
                       cmap='viridis', 
                       alpha=0.6, s=30)
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
ax2.set_title(f'Predicted Labels (KNN k={best_k})')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

misclassified = y_pred_final != y_test
n_misclassified = np.sum(misclassified)

print(f"\nTotal misclassified: {n_misclassified} out of {len(y_test)} ({n_misclassified/len(y_test)*100:.2f}%)")

if n_misclassified > 0:
    print("\nFirst 10 misclassifications:")
    print("-" * 60)
    misclass_indices = np.where(misclassified)[0][:10]
    for idx in misclass_indices:
        print(f"True: {y_test[idx]:15s} | Predicted: {y_pred_final[idx]:15s}")
```
